FAZENDO A ANÁLISE DE ENGAJAMENTO DAS PUBLICAÇÕES DO INSTAGRAM DE UMA LOJA

```python
# Importando biblioteca e criando conexão com o banco de dados
import pandas as pd 
import matplotlib.pyplot as plt
caminho_arquivo = r'C:\Projetos Fabio Damico\TudoPYTHON\Projetos_Ciencia-Dados\Pandas e Numpy\Instagram\08. Analisando o engajamento no Instagram.xlsx'
df = pd.read_excel(caminho_arquivo)
df.info()
df.head(15)
```

```python
# Retirando informações desnecessárias do banco de dados 
# Localizando as linhas sem valor para poder realizar uma lógica e compreender se é possível adicionar algum valor
# Fazendo leitura rápida da descrição e valores gerais da base 
df = df.drop('Visualizações', axis = 1)
df.loc[df.Carrossel.isnull()].head(8)
df.loc[df.Carrossel.isnull(),"Carrossel"] = "N"
df.loc[df.Tags.isnull(),"Tags"] = "Sem Tags"
#df.info()
df.describe()
#df.head(15)
```

```python
# Imprimindo os valores da base de acordo com os maiores números de curtidas
df.sort_values(by="Curtidas", ascending=False).head(10)
```

```python
# Imprimindo os valores da base de acordo com os menores números de curtidas
df.sort_values(by="Curtidas", ascending=False).tail(10)
```

```python
# Fazendo um "mini resumo" sobre qual tipo de publicação obtém maior curtidas e comentários, relacionando diretamente a pessoas
pd.options.display.float_format = '{:.2f}'.format
df.groupby(["Tipo","Pessoas"]) [["Curtidas","Comentários"]].mean()
```

```python
# Fazendo um "mini resumo" se campanhas S ou N obtém maior curtidas e comentários, relacionando diretamente a pessoas
df.groupby(["Pessoas","Campanhas"])[["Curtidas","Comentários"]].mean()
```

```python
# Verificando a média de curtidas por tipos de tags
df.groupby("Tags")["Curtidas"].mean()
```

```python
# Separando as Tags juntas por / e refazendo com , 
df.Tags = df.Tags.str.split('/')
df.head(5)
```

```python
# Afirmando qual o maior influenciador de sucesso dos posts
df.groupby(["Pessoas","Campanhas", "Carrossel"])[["Curtidas","Comentários"]].mean().sort_values("Curtidas",ascending=False)
```

```python
# Filtrar as publicações com pessoas e sem pessoas
pessoass = df[df['Pessoas'] == 'S']
pessoasn = df[df['Pessoas'] == 'N']

# Criar cópias dos dataframes filtrados com colunas relevantes
df_pessoass = pessoass[['Tipo', 'Pessoas', 'Curtidas', 'Comentários']].copy()
df_pessoasn = pessoasn[['Tipo', 'Pessoas', 'Curtidas', 'Comentários']].copy()

# Calcular médias de curtidas para cada tipo com pessoas
media_fotos_pessoass = df_pessoass[df_pessoass['Tipo'] == "Foto"]['Curtidas'].mean()
media_reels_pessoass = df_pessoass[df_pessoass['Tipo'] == "Reels"]['Curtidas'].mean()
media_videos_pessoass = df_pessoass[df_pessoass['Tipo'] == "Vídeo"]['Curtidas'].mean()

# Calcular médias de curtidas para cada tipo sem pessoas
media_fotos_pessoasn = df_pessoasn[df_pessoasn['Tipo'] == "Foto"]['Curtidas'].mean()
media_reels_pessoasn = df_pessoasn[df_pessoasn['Tipo'] == "Reels"]['Curtidas'].mean()
media_videos_pessoasn = df_pessoasn[df_pessoasn['Tipo'] == "Vídeo"]['Curtidas'].mean()

# Comparar médias de curtidas e imprimir resultados
if media_fotos_pessoass > media_fotos_pessoasn:
    tamanho_curt_fotos = media_fotos_pessoass / media_fotos_pessoasn
    print(f'As publicações do tipo -FOTO COM PESSOAS- possuem média de curtidas {media_fotos_pessoass:.2f} e comparado a publicações do tipo -FOTO SEM PESSOAS- é {tamanho_curt_fotos:.2f}x maior.')
    print("--"*72 )
else:
    tamanho_curt_foton = media_fotos_pessoasn / media_fotos_pessoass
    print(f'As publicações do tipo -FOTO SEM PESSOAS- possuem média de curtidas {media_fotos_pessoasn:.2f} e comparado a publicações do tipo -FOTO COM PESSOAS- é {tamanho_curt_foton:.2f}x maior.')
    print("--"*72 )

# Repetir o processo para os tipos de Reels
if media_reels_pessoass > media_reels_pessoasn:
    tamanho_curt_reels = media_reels_pessoass / media_reels_pessoasn
    print(f'As publicações do tipo -REELS COM PESSOAS- possuem média de curtidas {media_reels_pessoass:.2f} e comparado a publicações do tipo -REELS SEM PESSOAS- é {tamanho_curt_reels:.2f}x maior.')
    print("--"*72 )
else:
    tamanho_curt_reeln = media_reels_pessoasn / media_reels_pessoass
    print(f'As publicações do tipo -REELS SEM PESSOAS- possuem média de curtidas {media_reels_pessoasn:.2f} e comparado a publicações do tipo -REELS COM PESSOAS- é {tamanho_curt_reeln:.2f}x maior.')
    print("--"*72 )

# Repetir o processo para os tipos de Vídeo
if media_videos_pessoass > media_videos_pessoasn:
    tamanho_curt_videos = media_videos_pessoass / media_videos_pessoasn
    print(f'As publicações do tipo -VÍDEO COM PESSOAS- possuem média de curtidas {media_videos_pessoass:.2f} e comparado a publicações do tipo -REELS SEM PESSOAS- é {tamanho_curt_videos:.2f}x maior.')
    print("--"*72 )
else:
    tamanho_curt_videon = media_videos_pessoasn / media_videos_pessoass
    print(f'As publicações do tipo -VÍDEO SEM PESSOAS- possuem média de curtidas {media_videos_pessoass:.2f} e comparado a publicações do tipo -REELS COM PESSOAS- é {tamanho_curt_videon:.2f}x maior.')
    print("--"*72 )

# Calcular estatísticas de curtidas e comentários para publicações com pessoas
curtidas_pessoass = df_pessoass["Curtidas"].sum()
curtidas_pessoasn = df_pessoasn["Curtidas"].sum()
coment_pessoass = df_pessoass["Comentários"].sum()
coment_pessoasn = df_pessoasn["Comentários"].sum()
curtidas_tt = df["Curtidas"].sum()
coment_tt = df["Comentários"].sum()

# Calcular percentagens de curtidas e comentários para publicações com e sem pessoas
porc_come_pessoass = (coment_pessoass / coment_tt) * 100
porc_come_pessoasn = (coment_pessoasn / coment_tt) * 100
porc_curt_pessoass = (curtidas_pessoass / curtidas_tt) * 100
porc_pessoasn = (curtidas_pessoasn / curtidas_tt) * 100

# Imprimir resultados
print(f'As publicações COM PESSOAS possuem: {porc_curt_pessoass:.2f}% do TOTAL GERAL de curtidas, e {porc_come_pessoass:.2f}% dos comentários totais.')
print("--"*48 )
print(f'As publicações SEM PESSOAS possuem: {porc_pessoasn:.2f}% do TOTAL GERAL de curtidas, e {porc_come_pessoasn:.2f}% dos comentários totais.')
print("--"*48 )

# Comparar engajamento de curtidas e comentários para publicações com e sem pessoas
if curtidas_pessoass > curtidas_pessoasn:
    tamanho_curt = curtidas_pessoass / curtidas_pessoasn

    if coment_pessoass > coment_pessoasn:
        tamanho_come = coment_pessoass / coment_pessoasn
        print(f'As publicações COM pessoas se mostraram com engajamento de curtidas {tamanho_curt:.2f}x maior, e com o engajamento de comentários {tamanho_come:.2f}x maior, comparadas as publicações sem pessoas.')
        print("--"*85 )
    else:
        tamanho_come = coment_pessoasn / coment_pessoass
        print(f'As publicações COM pessoas se mostraram com engajamento de curtidas {tamanho_curt:.2f}x maior, e com o engajamento de comentários {tamanho_come:.2f}x menor, comparadas as publicações sem pessoas.')
        print("--"*85 )
else:
    tamanho_curt = curtidas_pessoasn / curtidas_pessoass

    if coment_pessoass > coment_pessoasn:
        tamanho_come = coment_pessoass / coment_pessoasn
        print(f'As publicações COM pessoas se mostraram com engajamento de curtidas {tamanho_curt:.2f}x menor, e com o engajamento de comentários {tamanho_come:.2f}x maior, comparadas as publicações com pessoas.')
        print("--"*85 )
    else:
        tamanho_come = coment_pessoasn / coment_pessoass
        print(f'As publicações COM pessoas se mostraram com engajamento de curtidas {tamanho_curt:.2f}x menor, e com o engajamento de comentários {tamanho_come:.2f}x menor, comparadas as publicações com pessoas.')
        print("--"*85 )
```
